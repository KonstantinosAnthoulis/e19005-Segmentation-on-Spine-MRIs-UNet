```python
csv_path = "C:/Users/kosta/Desktop/valid plot/valid_metrics.csv"

import pandas as pd

data_df = pd.read_csv(csv_path)  # Automatically handles headers

print((data_df.columns))

```

```python

# Extract relevant columns
epochs = data_df["epoch"]
avg_tloss = data_df["avg_tloss"]
avg_vloss = data_df["avg_vloss"]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(epochs, avg_tloss, label="Train Loss", marker="o", linestyle="-")
plt.plot(epochs, avg_vloss, label="Validation Loss", marker="s", linestyle="--")

# Labels and Title
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)

# Show Plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Function to extract float values from tensor-like strings
def extract_float(tensor_str):
    """Extracts the float value from a string representation of a PyTorch tensor."""
    match = re.search(r"[-+]?\d*\.\d+|\d+", str(tensor_str))  # Extract float or int
    return float(match.group()) if match else None  # Convert to float


data_df["avg_taccu"] = data_df["avg_taccu"].apply(extract_float)
data_df["avg_vaccu"] = data_df["avg_vaccu"].apply(extract_float)

# Extract relevant columns
epochs = data_df["epoch"]
avg_taccu = data_df["avg_taccu"]
avg_vaccu = data_df["avg_vaccu"]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(epochs, avg_taccu, label="Train Accuracy", marker="o", linestyle="-")
plt.plot(epochs, avg_vaccu, label="Validation Accuracy", marker="s", linestyle="--")

# Labels and Title
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()
plt.grid(True)

# Show Plot
plt.show()
```
